## Conceptual Exercises

### Exercise 1

(a) For each of parts (i) through (v), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

**Solution**:

- **(i) The sample size n is extremely large, and the number of predictors p is small**: A flexible method is better. With a large sample size, a flexible model can capture complex patterns without overfitting, as there is sufficient data to estimate the parameters accurately. Inflexible methods may underfit by failing to capture non-linear relationships.

- **(ii) The number of predictors p is extremely large, and the number of observations n is small**: An inflexible method is better. With a small sample size and many predictors, a flexible model is likely to overfit, capturing noise rather than true patterns. Inflexible methods, like linear regression, impose more structure and are less prone to overfitting.

- **(iii) The relationship between the predictors and response is highly non-linear**: A flexible method is better. Flexible models can capture non-linear relationships, while inflexible models (e.g., linear regression) may fail to model the complexity, leading to underfitting.

- **(iv) The variance of the error terms, i.e., σ² = Var(ε), is extremely high**: An inflexible method is better. High error variance means the data has a lot of noise. Flexible models may overfit by fitting to this noise, whereas inflexible models, with simpler assumptions, are more robust.

- **(v) The predictors are highly correlated with each other**: An inflexible method is better. High correlation among predictors can lead to instability in flexible models, as they may try to fit complex interactions that are not meaningful. Inflexible models, like those with regularization, can handle multicollinearity better.

(b) Explain whether each scenario in (i) through (v) would result in a high bias or high variance for a flexible statistical learning method, and why.

**Solution**:

- **(i) Large n, small p**: Low bias, low variance. A flexible model can fit complex patterns (low bias) and, with large n, has enough data to estimate parameters accurately, reducing variance.

- **(ii) Large p, small n**: Low bias, high variance. A flexible model can fit complex patterns (low bias) but, with small n, it overfits, leading to high variance as it captures noise.

- **(iii) Highly non-linear relationship**: Low bias, potentially high variance. A flexible model can capture the non-linear relationship (low bias), but without sufficient data, it may overfit, increasing variance.

- **(iv) High variance of error terms**: Low bias, high variance. A flexible model can fit the true relationship (low bias) but may also fit the high noise, leading to high variance.

- **(v) Highly correlated predictors**: Low bias, high variance. A flexible model can fit complex patterns (low bias), but correlated predictors can lead to unstable estimates, increasing variance.

### Exercise 2

Explain whether each scenario below is a classification, regression, or clustering problem, and indicate whether we are most interested in inference or prediction. Finally, provide n and p.

**Solution**:

- **(a) We collect a set of data on the top 500 firms in the US. For each firm we record profit, number of employees, industry and the CEO salary. We are interested in understanding which factors affect CEO salary.**
  - **Type**: Regression (CEO salary is a continuous response variable).
  - **Goal**: Inference (understanding which factors affect the response).
  - **n**: 500 (number of firms).
  - **p**: 3 (profit, number of employees, industry).

- **(b) We are considering launching a new product and wish to know whether it will be a success or a failure. We collect data on 20 similar products, recording their sales volume, various features of the product, the advertising budget, and whether it was a success or failure.**
  - **Type**: Classification (success or failure is a categorical response).
  - **Goal**: Prediction (predicting success or failure of the new product).
  - **n**: 20 (number of products).
  - **p**: Number of features (sales volume, product features, advertising budget; exact number depends on feature count, assume multiple features).

- **(c) We collect data on 1000 customers, including their purchase history, browsing history, demographics, and whether or not they have purchased a particular product. We wish to group customers into different types to better target marketing.**
  - **Type**: Clustering (grouping customers into types without a response variable).
  - **Goal**: Neither inference nor prediction (clustering aims to discover structure).
  - **n**: 1000 (number of customers).
  - **p**: Number of features (purchase history, browsing history, demographics; exact number depends on feature count).

### Exercise 3

This exercise relates to the College data set, which can be found in the file College.csv on the book website. It contains a number of variables for 777 colleges and universities in the US.

(a) Use the pd.read_csv() function to read the data into Python. Call the loaded data college. Make sure that you have the directory set to the correct location for the data.

(b) Look at the data used in the notebook by creating and running a new cell with just the code college in it. You should notice that the first column is just the name of each university in a column named something like Unnamed: 0. We don’t really want Python to treat this as data. However, it may be handy to have the university names as an index to the data. Use a command to set the index of college to this column.

(c) Use the describe() function to produce a numerical summary of the variables in the data set.

(d) Use pd.plotting.scatter_matrix() to produce a scatterplot matrix of the first ten columns or variables of the data. Recall that you can reference the first ten columns of a matrix A using A.iloc[:,:10].

(e) Use college.boxplot() to produce side-by-side boxplots of Outstate versus Private.

(f) Create a new qualitative variable, called Elite, by binning the Top10perc variable. We are going to divide universities into two categories based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%. Use pd.cut() for this. Then use college.boxplot() to produce side-by-side boxplots of Outstate versus Elite.

(g) Use college.hist() to produce some histograms with differing numbers of bins for a few of the quantitative variables. You may find the argument bins= useful.

(h) Continue exploring the data, and provide a brief summary of what you discover.

```python
import pandas as pd
import matplotlib.pyplot as plt

# (a) Load the data
college = pd.read_csv('data/College.csv')

# (b) Set the index to the university names
college = college.set_index('Unnamed: 0')
college.index.name = 'Name'

# (c) Numerical summary
print(college.describe())

# (d) Scatterplot matrix of first ten columns
pd.plotting.scatter_matrix(college.iloc[:, :10], figsize=(10, 10))
plt.show()

# (e) Boxplot of Outstate vs Private
college.boxplot(column='Outstate', by='Private')
plt.title('Outstate Tuition by Private Status')
plt.suptitle('')
plt.show()

# (f) Create Elite variable and boxplot
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])
college.boxplot(column='Outstate', by='Elite')
plt.title('Outstate Tuition by Elite Status')
plt.suptitle('')
plt.show()

# (g) Histograms with different bins
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
college['Apps'].hist(bins=10, ax=axes[0, 0]).set_title('Applications (10 bins)')
college['Apps'].hist(bins=30, ax=axes[0, 1]).set_title('Applications (30 bins)')
college['Enroll'].hist(bins=10, ax=axes[1, 0]).set_title('Enrollment (10 bins)')
college['Enroll'].hist(bins=30, ax=axes[1, 1]).set_title('Enrollment (30 bins)')
plt.tight_layout()
plt.show()

# (h) Summary of findings
print('Summary of findings:')
print('- The dataset contains 777 universities with various features like applications, enrollment, tuition, and academic metrics.')
print('- Private universities tend to have higher out-of-state tuition compared to public ones, as seen in the boxplot.')
print('- Elite universities (Top10perc > 50%) have significantly higher out-of-state tuition than non-elite ones.')
print('- Histograms show that variables like Applications and Enrollment are right-skewed, with a few universities having very high values.')
print('- The scatterplot matrix suggests potential correlations between variables like Apps and Accept, or Enroll and Accept, which could be explored further.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
boston = pd.read_csv('data/Boston.csv')

# (a) Data dimensions
print(f'(a) Rows: {boston.shape[0]}, Columns: {boston.shape[1]}')
print('Rows represent individual suburbs (towns) in Boston. Columns represent predictors/features like crime rate, tax rate, etc., and the response variable (median home value).')

# (b) Pairwise scatterplots
pd.plotting.scatter_matrix(boston[['crim', 'zn', 'indus', 'nox', 'rm', 'age', 'dis']], figsize=(10, 10))
plt.show()
print('Findings: Scatterplots show relationships like a positive correlation between rm (average rooms) and house prices, and a negative correlation between dis (distance to employment centers) and nox (nitrogen oxides concentration). Some variables like crim (crime rate) are highly skewed.')

# (c) Predictors associated with crime rate
print('(c) Correlation with crime rate:')
print(boston.corr()['crim'].sort_values())
print('Crime rate (crim) is positively correlated with indus (proportion of non-retail business), nox (nitrogen oxides), and rad (accessibility to highways), and negatively correlated with dis (distance to employment centers) and medv (median home value). High crime rates tend to occur in areas with more industrial activity and pollution, and lower home values.')

# (d) High crime rates, tax rates, pupil-teacher ratios
print('(d) Summary statistics:')
print(boston[['crim', 'tax', 'ptratio']].describe())
print('High crime rates: Some suburbs have crime rates as high as 88.98 (max), with the 75th percentile at 3.68, indicating a few outliers with very high crime.')
print('High tax rates: The tax rate ranges from 187 to 711, with some suburbs having very high tax rates (max 711).')
print('Pupil-teacher ratios: The ratio ranges from 12.6 to 22, with most suburbs between 16.1 and 20.2 (25th to 75th percentile).')

# (e) Suburbs bounding the Charles river
print(f'(e) Suburbs bounding Charles river: {boston["chas"].sum()}')

# (f) Median pupil-teacher ratio
print(f'(f) Median pupil-teacher ratio: {boston["ptratio"].median()}')

# (g) Suburb with lowest median home value
lowest_medv = boston.loc[boston['medv'].idxmin()]
print('(g) Suburb with lowest median home value:')
print(lowest_medv)
print('Comparison to ranges:')
print(boston.describe())
print('The suburb has a high crime rate (38.35, above 75th percentile), low zn (0, min), high indus (18.1, above median), high nox (0.693, above median), low rm (5.453, below median), high age (91.8, above 75th percentile), and low dis (1.42, below 25th percentile). This suggests a less desirable area with high crime and industrial activity.')

# (h) Suburbs with more than 7 or 8 rooms
print(f'(h) Suburbs with >7 rooms: {len(boston[boston["rm"] > 7])}')
print(f'Suburbs with >8 rooms: {len(boston[boston["rm"] > 8])}')
print('Suburbs with >8 rooms tend to have higher median home values (medv) and lower crime rates, indicating they are likely wealthier, more desirable areas.')
```
